```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['IV2 TF','IV3 Torch', 'IV3 Keras']
data_1 = [0.756, 0.772, 0.777]
data_2 = [2.341, 2.718, 2.736]
data_3 = [5.429,6.216,5.917]

# Set up plot
x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(figsize=(10, 8))
rects1 = ax.bar(x - width, data_1, width, label='ImageNet', color='#1f77b4')
rects2 = ax.bar(x, data_2, width, label='COCO', color='#ff7f0e')
rects3 = ax.bar(x + width, data_3, width, label='CIFAR10', color='#2ca02c')

# Add labels, title, and legend
ax.set_xlabel("Networks")
ax.set_ylabel('Inception Score')
ax.set_title('Inception Score Comparisons')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Show plot
plt.savefig("Inception Score.png")
```
